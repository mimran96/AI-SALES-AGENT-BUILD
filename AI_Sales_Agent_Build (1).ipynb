# Configure Deepgram for Speech-to-Text (STT)

```python
# Import Libraries
import os
from deepgram import Deepgram
import asyncio
import aiohttp
```

```python
# This function will process audio files and transcribe them into text.
async def transcribe_audio(audio_url):
    async with aiohttp.ClientSession() as session:
        deepgram = Deepgram(DEEPGRAM_API_KEY)

        response = await deepgram.transcription.prerecorded(
            { "url": audio_url },
            { "smart_format": True }
        )

        transcript = response["results"]["channels"][0]["alternatives"][0]["transcript"]
        return transcript
```

```python
AUDIO_URL = "https://www.bruntwork.co/wp-content/uploads/2023/03/BruntWork-customer-service-agent-sample-1.mp3"

transcription = asyncio.create_task(transcribe_audio(AUDIO_URL))
print("Transcription:", transcription)
```

```python
import vocode
from vocode.streaming.models.transcriber import DeepgramTranscriberConfig
from vocode.streaming.models.synthesizer import ElevenLabsSynthesizerConfig
from vocode.streaming.models.agent import ChatGPTAgentConfig
from vocode.streaming.telephony.server.base import TelephonyServer
from vocode.streaming.agent.base_agent import BaseAgent
from vocode.streaming.synthesizer.eleven_labs_synthesizer import ElevenLabsSynthesizer
from vocode.streaming.transcriber.deepgram_transcriber import DeepgramTranscriber
import asyncio
```

```python
from vocode.streaming.models.agent import AgentConfig

class MistralSalesAgent(BaseAgent):
    def __init__(self):
        super().__init__(AgentConfig(type="custom"))  # Fix: Add required config
        self.history = []

    async def process_message(self, text: str) -> str:
        self.history.append(f"User: {text}")
        prompt = "\n".join(self.history) + "\nAI Sales Agent:"

        inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
        outputs = model.generate(**inputs, max_new_tokens=100)
        response = tokenizer.decode(outputs[0], skip_special_tokens=True)

        self.history.append(f"AI Sales Agent: {response}")
        return response
```

```python
# Configure Vocode to Handle Real-Time Conversations
def create_vocode_agent():

    transcriber = DeepgramTranscriber(DeepgramTranscriberConfig(
        sampling_rate=16000,  # Set appropriate sampling rate
        audio_encoding="linear16",  # Supported encoding: "linear16", "opus", etc.
        chunk_size=1024,  # Adjust as needed
        api_key=DEEPGRAM_API_KEY))

    synthesizer = ElevenLabsSynthesizer(
        ElevenLabsSynthesizerConfig(api_key=ELEVENLABS_API_KEY,
                                    voice="Aria",
                                    sampling_rate=22050,  # Set appropriate sampling rate
                                    audio_encoding="linear16",  # Supported encoding: "mp3", "wav")
                                ))

    agent = MistralSalesAgent()

    return transcriber, synthesizer, agent
```

```python
from vocode.streaming.telephony.server.base import TelephonyServer
from vocode.streaming.models.telephony import TwilioConfig

# Function to start the server properly
async def start_sales_agent():
    print("Hello")
    transcriber, synthesizer, agent = create_vocode_agent()

    server = TelephonyServer(
        transcriber_config=DeepgramTranscriberConfig(api_key=DEEPGRAM_API_KEY),
        agent_config=AgentConfig(type="custom"),
        synthesizer_config=ElevenLabsSynthesizerConfig(api_key=ELEVENLABS_API_KEY, voice="Aria"),
        twilio_config=TwilioConfig(
            account_sid=TWILIO_ACCOUNT_SID,
            auth_token=TWILIO_AUTH_TOKEN,
            twilio_phone_number=TWILIO_PHONE_NUMBER
        )
    )

    server.run()
```
